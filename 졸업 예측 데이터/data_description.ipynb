```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
#df = pd.read_csv("C:\\Users\\yoona\\Desktop\\std_data\\*.csv" , encoding='UTF8')
file_path_list = glob.glob("C:\\Users\\yoona\\Desktop\\std_data\\*.csv")
tong = pd.DataFrame(columns=['학생소속대학명', '학생교육부계열', 'gpa'])
for file_path in file_path_list:
    data = pd.read_csv(file_path, encoding='UTF8')
    need = data[['학생소속대학명', '학생교육부계열', 'gpa']]
    tong = pd.concat([tong,need])

tong
```

```python
import seaborn as sns
# 한글 폰트 적용
plt.rc('font', family='Malgun Gothic') 
# 캔버스 사이즈 적용
plt.rcParams["figure.figsize"] = (50, 20)

sns.countplot(x="학생소속대학명", data=tong, order = tong['학생소속대학명'].value_counts().index)
plt.show()
```

```python
tong['학생소속대학명'].value_counts()['공과대학'] / tong['학생소속대학명'].count()
```

```python
tong['학생교육부계열'].value_counts()['인문·사회'] / tong['학생교육부계열'].count()
```

```python
plt.rcParams["figure.figsize"] = (10, 6)
sns.set_palette("pastel")
sns.countplot(x="학생교육부계열", data=tong, order = tong['학생교육부계열'].value_counts().index)

plt.show()
```
